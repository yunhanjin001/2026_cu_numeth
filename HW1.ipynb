The code below is an example of using Brent's method

```python
import numpy as np
import scipy.optimize as spop
import timeit
```

```python
# NPV function. tk and ck are the array of time and cachflow.
def npv(r, tk, ck):
  assert(len(tk) == len(ck))
  return np.sum(np.power(1.0 + r, -tk) * ck)
```

```python
#f(r)
def root(r, p, tk, ck):
  return npv(r, tk, ck) - p
```

```python
# f'(r)
def droot(r, tk, ck):
    # d(1+r)^(-tk) = -tk*(1+r)^(-tk-1)
    return np.sum(-tk * np.power(1.0 + r, -tk - 1) * ck)
```

```python
# Newton
def newton_method(p, tk, ck, r0=0.05, tol=1e-8, max_iter=100):
    r = r0
    for i in range(max_iter):
        f = root(r, p, tk, ck)
        df = droot(r, tk, ck)

        # avoid division by zero
        if abs(df) < 1e-12:
            raise RuntimeError("Derivative too small")

        r_new = r - f / df

        if abs(r_new - r) < tol:
            return r_new

        r = r_new

    raise RuntimeError("Newton did not converge")
```
