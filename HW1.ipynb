Implement the newton's metehod to compute the implied yield.

The code below is an example of using Brent's method

```python
import numpy as np
import scipy.optimize as spop
import timeit
```

```python
# NPV function. tk and ck are the array of time and cachflow.
def npv(r, tk, ck):
  assert(len(tk) == len(ck))
  return np.sum(np.power(1.0 + r, -tk) * ck)
```

```python
#f(r)
def root(r, p, tk, ck):
  return npv(r, tk, ck) - p
```

```python
rng = np.random.default_rng(12345)
```

```python
%%time
for k in range(1000):
  # Randomly generating cashflow, time, and price
  ck = rng.uniform(0, 10, 100)
  ck[-1] += 100
  tk = np.cumsum(rng.uniform(0.1, 0.3, 100))
  p = rng.uniform(10.0, 100)

  args = (p, tk, ck)
  r = spop.newton(root, 0.0, args=args, maxiter=20)
  assert(np.abs(root(r, p, tk, ck))<1e-8)
```

1. We can make a good approcimation by:
    r<sub>0</sub> = (Σ<sup>t</sup> c<sub>k</sub> / p)<sup>1/t</sup>
 - r<sub>0</sub> is the start point.
 - Σ<sup>t</sup> c<sub>k</sub> is the total amount expected to receive when discount rate is 0.
 - t is the expected time to have cash flow back.


 2. P(r) is decreasing. P'(r) <0.

    P'(r)=Σck(-tk)(1+r)<sup>-tk-1</sup>

    P"(r)=Σck(-tk)(-tk-1)(1+e) <sup>-tk-2</sup>  >0

    So P(r) is decreasing and convex.
